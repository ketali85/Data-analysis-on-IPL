```python
deliveries=pd.read_csv("C:/Ketali/excel/deliveries.csv")
deliveries.head()
```

<b><h2>Batsman analysis:</h2></b>


Let us start our analysis with batsman. Let us first see the ones with most number of IPL runs under their belt.

```python
#
df1 = deliveries.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
df1 = df1.iloc[:10]
df1
labels = np.array(df1['batsman'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,7))
rects = ax.bar(ind, np.array(df1['batsman_runs']), width=width, color='plum')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')

ax.set_title("Top run scorers in IPL")
ax.set_xlabel('Batsmane Name')

plt.show()
```

<b>Virat Kohli </b> is leading the chart followed closely by Raina. Gayle is the top scorer among foreign players.

```python
# Now let us see the players with more number of boundaries in IPL.
df2 = deliveries.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
df2 = df2.iloc[:10]

labels = np.array(df2['batsman'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(df2['batsman_runs']), width=width, color='wheat')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Batsman with most number of boundaries.!",fontsize = 10)

plt.show()
```

 <b>Gambhir</b> is way ahead of others - almost 60 boundaries more than <b>Kohli</b>.! Nice to <b> Sachin </b> in the top 10 list 

```python

# Now let us check the number of 6's
df3 = deliveries.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
df3 = df3.iloc[:10]

labels = np.array(df3['batsman'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(df3['batsman_runs']), width=width, color='slategrey')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation=90)
ax.set_ylabel("Count")
ax.set_title("Batsman with most number of sixes.!")
ax.set_xlabel('Batsmane Name')

plt.show()

```

<b>Gayle</b>, the unassailable leader in the number of sixes. <br><b>Raina</b> is third in both number of 4's and 6's</br>

<b> <h2>Bowler analysis </h2></b>


Now let us see the bowlers who has bowled most number of balls in IPL.



```python
ball1 = deliveries.groupby('bowler')['ball'].agg('count').reset_index().sort_values(by='ball', ascending=False).reset_index(drop=True)
ball1 = ball1.iloc[:10,:]

labels = np.array(ball1['bowler'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(ball1['ball']), width=width, color='mediumpurple')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers - Number of balls bowled in IPL")
ax.set_xlabel('Bowler Names')

plt.show()
```

<b>Harbhajan Singh</b> is the the bowler with most number of balls bowled in IPL matches. Now let us see the<br>
bowler with more number of dot balls.</br>

```python
ball2 = deliveries.groupby('bowler')['total_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='total_runs', ascending=False).reset_index(drop=True)
ball2 = ball2.iloc[:10,:]

labels = np.array(ball2['bowler'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(ball2['total_runs']), width=width, color='palevioletred')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers - Number of dot balls bowled in IPL")
ax.set_xlabel('Bowler Names')

plt.show()
```

<b>Pravin Kumar</b> is the one with more number of dot balls followed by <b>Steyn</b> and <b>Malinga</b>



```python
# Now let us see the bowlers who has bowled more number of extras in IPL.
ball3 = deliveries.groupby('bowler')['extra_runs'].agg(lambda x: (x>0).sum()).reset_index().sort_values(by='extra_runs', ascending=False).reset_index(drop=True)
ball3 = ball3.iloc[:10,:]

labels = np.array(ball3['bowler'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(ball3['extra_runs']), width=width, color='pink')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Bowlers with more extras in IPL")
ax.set_xlabel('Bowler Names')

plt.show()
```

<b>Malinga</b> tops the chart with 221 extra runs followed by <b>Pravin Kumar.</b>
